```python
# Import a bunch of libraries.
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# This tells matplotlib not to try opening a new window for each plot.
%matplotlib inline

# Generally you should only filter specific classes of warnings if you are sure
# you know what you're doing!

# import warnings
# warnings.filterwarnings('ignore')
```

Create some artificial data. Try changing the seed, n_samples, and degrees.

```python
# Set the randomizer seed so results are the same each time.
np.random.seed(100)

# How many samples to generate.
n_samples = 20

# Generate evenly spaced X values in [0,1] using linspace.
X = np.linspace(0, 1, n_samples)

# Set the true function as a piece of a cosine curve.
true_fun = lambda X: np.cos(1.5 * np.pi * X)

# Generate true y values. Then add random noise.
y = true_fun(X)
y += np.random.randn(n_samples) * 0.2

# Below, we'll fit polynomials to the noisy data with these degrees.
degrees = [1, 4, 15]
```

```python
# Initialize a new plot.
plt.figure(figsize=(14, 4))

# We'll create a subplot for each value of the degrees list (see above).
for i in range(len(degrees)):
    # The subplots are all on the same row.
    ax = plt.subplot(1, len(degrees), i+1)
    
    # Turn off tick marks to keep things clean.
    plt.setp(ax, xticks=(), yticks=())

    # We achieve polynomial regression by performing linear regression over all polynomial combinations of the input features.
    # In this case, there is only a single input feature, the value of x, so the degree k features include x, x^2, ..., x^k.
    polynomial_features = PolynomialFeatures(degree=degrees[i],
                                             include_bias=False)

    # Use the sklearn's Pipeline to string together 2 operations.
    linear_regression = LinearRegression()
    pipeline = Pipeline([("polynomial_features", polynomial_features),
                         ("linear_regression", linear_regression)])

    # X is an array with n_samples values. But the fit() function expects an array of feature vectors -- an array of arrays:
    # [[f1], [f2], ...]. Another way to do this is np.transpose([X]). Read more about array indexing for details.
    pipeline.fit(X[:, np.newaxis], y)
    
    # Show samples from the fitted function.
    X_test = np.linspace(0, 1, 100)
    plt.plot(X_test, pipeline.predict(X_test[:, np.newaxis]), label="Model")

    # Show the true function.
    plt.plot(X_test, true_fun(X_test), label="True function")

    # Show the original noisy samples.
    plt.scatter(X, y, label="Samples")

    # Add a few more labels to the plot.
    plt.xlabel("x")
    plt.ylabel("y")
    plt.xlim((-.05, 1.05))
    plt.ylim((-2, 2))
    plt.legend(loc="best")
    plt.title("Degree %d" % degrees[i])

# Render the plots.
plt.show()
```
